```python
county = []
area = []
lat = []
lon = []
content = []
for folder1 in os.listdir("txt_files"):
    for folder2 in os.listdir(os.getcwd()+"\\txt_files\\"+folder1):
        for file in os.listdir(os.getcwd()+"\\txt_files\\"+folder1+"\\"+folder2):
            file_name = os.getcwd()+"\\txt_files\\"+folder1+"\\"+folder2+"\\"+file
            with open(file_name,"r", encoding="utf8") as fh:
                file_text = fh.read()
                for para in file_text.split("\n\n\n"):
                    content.append(para)
                    area.append(" ".join(folder2.split("_")[:-2]))
                    lat.append(folder2.split("_")[-2])
                    lon.append(folder2.split("_")[-1])
                    county.append(folder1)
                    #print(content)
                    #print(area,lat,lon,county)
                
```

```python
df = pd.DataFrame({"county":county, "area":area, "lat":lat, "lon":lon, "content":content})
```

```python
df
```

```python
import spacy
```

```python
nlp = spacy.load("en_core_web_sm")
```

```python
entities = dict()
count = 0
for para in df["content"]:
    doc = nlp(para)
    for ent in doc.ents:
        # Print the entity text and its label
        count = count + 1
        print(ent.text, ent.label_)
        entities[ent.text] = entities.get(ent.text, 0) + 1
```
